```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()

# Database
captured_data = pd.read_csv('CholeraData.csv')

# Display Data
captured_data

```

```python
# Plotting the Bar chart
plt.figure(figsize=(9, 6))
bars = plt.bar(x=captured_data['Country'], height=captured_data['Reported Cases'])
plt.xlabel('Country')
plt.ylabel('Reported Cases')
plt.title('Reported Cases by Country')
plt.xticks(rotation=45)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.show()
```
